# Demographic Filtering

recommend best movie to users

```python
c=df['vote_average'].mean()
c
```

```python
m=df['vote_count'].quantile(0.8)
m
```

```python
popular=df[df['vote_count']>m].copy()
```

```python
def weighted_rating(x):
    v=x['vote_count']
    r=x['vote_average']
    return ((v*r)+(m*c))/(v+m)
```

```python
popular['score']=popular.apply(weighted_rating,axis=1)
```

```python
popular=popular.sort_values(by='score',ascending=False)
```

# Content based filtering

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['overview'].isnull().value_counts()
```

```python
df['overview'].fillna(" ",inplace=True)
```

```python
tfid=TfidfVectorizer(stop_words='english')
```

```python
tfid_matrix=tfid.fit_transform(df['overview'])
```

```python
tfid_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim=linear_kernel(tfid_matrix, tfid_matrix)
cosine_sim.shape
```

```python
from ast import literal_eval
features = ['cast','crew','keywords','genres']
for feature in features:
    df[feature] = df[feature].apply(literal_eval)
```

```python
df[features].isnull().value_counts()
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return ""
```

```python
df['Director']=df['crew'].apply(get_director)
```

```python
def top3(x):
    if isinstance(x,list):
        names=[i['name'] for i in x]
        if len(names)>3:
            return names[:3]
        return names
    return []
```

```python
features=['cast','keywords','genres']
for feature in features:
    df[feature]=df[feature].apply(top3)
```
